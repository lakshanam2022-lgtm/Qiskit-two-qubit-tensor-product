```python
# Task 1: Change Gate Combination
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# Create simulator
simulator = AerSimulator()

# Create a 2-qubit circuit
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.h(1)
qc.measure([0, 1], [0, 1])

# Transpile and run
compiled = transpile(qc, simulator)
job = simulator.run(compiled, shots=1024)
result = job.result()
counts = result.get_counts()

# Print only textual output
print("Output probabilities:", counts)
print("Observation: Both qubits are in superposition — uniform distribution across 00, 01, 10, and 11.")
```

```python
# Task 2: Swap Qubit Roles
qc = QuantumCircuit(2, 2)
qc.x(1)
qc.measure([0, 1], [0, 1])

t_qc = transpile(qc, simulator)
result = simulator.run(t_qc, shots=1024).result()
counts = result.get_counts()

print("X gate on second qubit output:", counts)
print("Compared to original circuit, only second qubit flips (state 00 → 01).")
```

```python
# Task 3: Add a CNOT Gate
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

t_qc = transpile(qc, simulator)
result = simulator.run(t_qc, shots=1024).result()
counts = result.get_counts()

print("CNOT after Hadamard output:", counts)
print("This creates an entangled Bell state (00 and 11 have ~equal probability).")
```

```python
# Task 4: Increase Measurement Shots
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

t_qc = transpile(qc, simulator)
result = simulator.run(t_qc, shots=4096).result()
counts_high = result.get_counts()

print("Output with 4096 shots:", counts_high)
print("Distribution becomes more stable and closer to 50–50 for 00 and 11.")
```

```python
# ✅ Final Task 5 Code (with guaranteed visible output)

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
from IPython.display import display  # <-- ensures notebook shows diagrams

# --- 1. Create circuit (Hadamard on both qubits) ---
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)

# --- 2. Visualize Bloch multivector BEFORE measurement ---
state = Statevector.from_instruction(qc)
display(plot_bloch_multivector(state))   # shows Bloch sphere

# --- 3. Add measurement and run simulation ---
qc.measure_all()
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()
counts = result.get_counts()

# --- 4. Display circuit and histogram ---
display(qc.draw("mpl"))                 # circuit diagram
display(plot_histogram(counts))         # histogram
```
